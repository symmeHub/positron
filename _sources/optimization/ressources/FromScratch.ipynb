```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
from data import gen_data
```

```python
x, y = gen_data()
x, y
```

```python
# plot
plt.figure()
plt.scatter(x, y)
```

```python
def F(x, a = 1, y0=0, b = 1):
    rep = y0 +np.sqrt(a*x+b**2*x**2)
    return rep

y_rep = F(x ,y0=0.1, a=0.2, b=0.1)
plt.figure()
plt.scatter(x, y)
plt.plot(x, y_rep, 'r-')
plt.show()
```

```python
def fc(P, x):
    y0 = P[0]
    a = P[1]
    b = P[2]
    y_aprox = F(x, y0=y0, a=a, b=b)
    res = (y_aprox - y)**2
    res = np.sum(res)
    return np.sqrt(res)/len(x)

P = np.array([0.1, 0.2, 0.1])
fc(P, x)
```

```python
from scipy.optimize import minimize
P_init = np.array([0.5, 0.5, 0.4])
res = minimize(fc, P_init, args=(x,))
p_opt = res.x
p_opt
```

```python
plt.figure()
plt.scatter(x, y)
#p_init
plt.plot(x, F(x, y0=P_init[0], a=P_init[1], b=P_init[2]), 'g-')
#p_opt
plt.plot(x, F(x, y0=p_opt[0], a=p_opt[1], b=p_opt[2]), 'r-')
plt.show()
```
